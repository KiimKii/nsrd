```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from konlpy.tag import Okt
from collections import Counter
from wordcloud import WordCloud
from PIL import Image
```

```python
df1 = pd.read_csv('아이폰 리뷰1.csv', encoding = 'cp949')
```

```python
df1["text"] = df1["text1"] + df1["text2"] + df1["text3"]
```

```python
df2 = pd.read_csv('아이폰 리뷰2.csv', encoding = 'cp949')
df3 = pd.read_csv('아이폰 리뷰3.csv', encoding = 'cp949')
df4 = pd.read_csv('아이폰 리뷰4.csv', encoding = 'cp949')
df5 = pd.read_csv('아이폰 리뷰5.csv', encoding = 'cp949')
df6 = pd.read_csv('아이폰 리뷰6.csv', encoding = 'cp949')
df7 = pd.read_csv('아이폰 리뷰7.csv', encoding = 'cp949')
df8 = pd.read_csv('아이폰 리뷰8.csv', encoding = 'cp949')
df9 = pd.read_csv('아이폰 리뷰9.csv', encoding = 'cp949')
df10 = pd.read_csv('아이폰 리뷰10.csv', encoding = 'cp949')
```

```python
df2["text"] = df2["text1"] + df2["text2"]
df3["text"] = df3["text1"]
df4["text"] = df4["text1"] + df4["text2"]
df5["text"] = df5["text1"] + df5["text2"]
df6["text"] = df6["text1"] + df6["text2"]
df7["text"] = df7["text1"] + df7["text2"]
df8["text"] = df8["text1"] + df8["text2"]
df9["text"] = df9["text1"] + df9["text2"]
df10["text"] = df10["text1"] + df10["text2"]
```

```python
df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10], ignore_index=True)
```

```python
twitter = Okt()
```

```python
df00 = df.dropna(axis = 0)
```

```python
word_list = []
word_list = df00['text']
```

```python
sentences_tag = []

for sentence in word_list:
    morph = twitter.pos(sentence)
    sentences_tag.append(morph)
    
print(morph)
```

```python
noun_list = []
for sentence in sentences_tag:
    for word, tag in sentence:
        if tag in ['Noun']:
            noun_list.append(word)
```

```python
len(noun_list)

noun_list = [n for n in noun_list if len(n) > 1]
noun_list[:100]
```

```python
counts = Counter(noun_list)
tags = counts.most_common(30)
tags
```

```python
wordcloud = WordCloud(font_path='C:/Windows/Fonts/malgun.ttf', background_color = 'white',
                      relative_scaling = 0.2).generate_from_frequencies(dict(tags))

print(dict(tags))


plt.figure(figsize=(10, 8))
plt.axis('off')
plt.imshow(wordcloud)
plt.show()
```
