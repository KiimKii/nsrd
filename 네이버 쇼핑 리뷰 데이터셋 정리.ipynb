```python
import pandas as pd
import numpy as np
import urllib.request
```

# 데이터 로드

```python
#1 데이터 로드
urllib.request.urlretrieve("https://raw.githubusercontent.com/bab2min/corpus/master/sentiment/naver_shopping.txt", filename="naver_shopping.txt")
```

```python
total_data = pd.read_table('naver_shopping.txt', names=['ratings', 'reviews'])
```

```python
total_data[:5]
```

# 데이터 전처리

## 레이블 만들기

```python
#3 레이블 만들기
total_data['label'] = np.select([total_data.ratings > 2], [1], default=0)
total_data
```

## 중복 및 null 값 제거

```python
#4 중복된 샘플 제거
total_data['ratings'].nunique(), total_data['reviews'].nunique(), total_data['label'].nunique()     # 확인
```

```python
total_data.drop_duplicates(subset=['reviews'], inplace=True)     # reviews 열에서 중복 내용 있다면 제거
print('총 샘플의 수 :',len(total_data))
```

```python
print(total_data.isnull().values.any())   # null값 확인
```

# 데이터 저장

```python
#5 데이터 저장
total_data.to_csv('nsrd.txt',index=False, sep='\t') # 구분자 탭으로 하여 저장. 인덱스칼럼 저장 안함.
```
